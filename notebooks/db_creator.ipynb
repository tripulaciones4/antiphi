```python
query1 = '''DROP TABLE IF EXISTS web_table'''
cursor.execute(query1)

query = '''CREATE TABLE web_table (url TEXT(1000), status VARCHAR(255))'''
cursor.execute(query)
```

```python
cols = ",".join([str(i) for i in df_reduced.columns.tolist()])
counter = 0
for i,row in df_reduced.iterrows():
    sql = "INSERT INTO web_table (" +cols + ") VALUES (" + "%s,"*(len(row)-1) + "%s)"
    cursor.execute(sql, tuple(row))
    counter = counter + 1
    print(counter)

    # the connection is not autocommitted by default, so we must commit to save our changes
    connection.commit()
```

```python
cursor.execute("SELECT * FROM web_table")

myresult = cursor.fetchall()

for x in myresult:
  print(x)
```
